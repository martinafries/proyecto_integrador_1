# Análisis de Ventas de Tecnología

Este notebook analiza los datos de ventas de productos tecnológicos durante el año 2023.

```python
# Importar librerías necesarias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Cargar datos
# Reemplaza la URL con la de tu archivo raw en GitHub
url = 'https://raw.githubusercontent.com/martinafries/proyecto_integrador_1/refs/heads/main/ventas_tecnologia.csv'
df = pd.read_csv(url)
df.head()
```

```python
# Crear gráfico de líneas para todas las categorías
plt.figure(figsize=(15, 8))


plt.xlabel('Mes')
plt.ylabel('Unidades Vendidas')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# prompt: generar datos descriptivos

# Estadísticas descriptivas básicas
print(df.describe())

# Distribución de las ventas por categoría
print(df.groupby('Categoria')['Unidades Vendidas'].describe())

# Correlación entre variables
correlation_matrix = df.corr()
print(correlation_matrix)

# Gráfico de barras de la distribución de ventas por categoría
plt.figure(figsize=(10, 6))
sns.countplot(x='Categoria', data=df)
plt.title('Distribución de Ventas por Categoría')
plt.xlabel('Categoría')
plt.ylabel('Número de Unidades Vendidas')
plt.xticks(rotation=45)
plt.show()

```
